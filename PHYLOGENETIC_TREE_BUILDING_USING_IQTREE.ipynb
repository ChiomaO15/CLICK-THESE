# Phylogenetic tree construction of *Lactiplantibacillus plantarum* and *Levilactobacillus brevis*

This repository contains the pipeline to construct a Maximum-Likelihood (ML) phylogenetic tree based on single-copy core genes. The analysis compares seven novel study isolates against representative *Lactiplantibacillus plantarum* and *Levilactobacillus brevis* genomes.

## Tree Specifications
- **Data Source**: Single-copy core genes extracted from 7 study isolates and public reference genomes.
- **Tree Type**: Maximum-Likelihood (ML).
- **Outgroup**: *Pediococcus pentosaceus*.

### 1. Extract and Align Core Genes
Run the alignment script on your isolated single-copy core genes.I did that on another code page and downloaded it and uploaded here.Proceed in building the tree with the alignment fasta


```python
!iqtree2 -s /content/core_alignment.fasta \
         -m MFP \
         -bb 1000 \
         -alrt 1000 \
         -nt AUTO \
         -o Pediococcus_pentosaceus_SRCM102740
```

### 2. Download the IQTREE File
After the tree is constructed,download the output with the tree file and proceed to view on ITOL or use Phylo and Matplotlib to view using the below script.

```python
from Bio import Phylo

tree_path = "/content/core_alignment.fasta.treefile"
tree = Phylo.read(tree_path, "newick")

# List all tip names
tip_names = [term.name for term in tree.get_terminals()]
print("Tip labels in the tree:")
for name in tip_names:
    print(name)
```

```python
from Bio import Phylo

# Load the tree
tree_path = "/content/core_alignment.fasta.treefile"
tree = Phylo.read(tree_path, "newick")

# Reroot on the correct outgroup
tree.root_with_outgroup("Pediococcus_pentosaceus_SRCM102740")

# Save the rerooted tree
rerooted_path = "/content/core_alignment_rerooted.nwk"
Phylo.write(tree, rerooted_path, "newick")

# Display a simple ASCII version
Phylo.draw_ascii(tree)
```
